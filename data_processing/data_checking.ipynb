```python
import os
from PIL import Image
import matplotlib.pyplot as plt

def inspect_images(directory):
    sizes = {}
    corrupted = []
    for subdir, dirs, files in os.walk(directory):
        for file in files:
            filepath = os.path.join(subdir, file)
            try:
                with Image.open(filepath) as img:
                    size = f"{img.size[0]}x{img.size[1]}"  # Store size as 'widthxheight' string
                    sizes[size] = sizes.get(size, 0) + 1
            except (IOError, SyntaxError) as e:
                corrupted.append(filepath)
    
    print("Image Sizes Distribution:")
    for size, count in sizes.items():
        print(f"Size {size} : {count} images")
    
    print("\nCorrupted Files:")
    for file in corrupted:
        print(file)

    # Plot the distribution of image sizes
    sizes_sorted = sorted(sizes.items())  # Sort sizes for better visualization
    labels, values = zip(*sizes_sorted)   # Unzip keys and values
    indexes = range(len(labels))
    plt.figure(figsize=(10, 5))
    plt.bar(indexes, values, color='blue')
    plt.xticks(indexes, labels, rotation='vertical')
    plt.xlabel('Image Size (width x height)')
    plt.ylabel('Frequency')
    plt.title('Distribution of Image Sizes')
    plt.show()
```

```python
inspect_images('dataset')
```
